```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.scatter(df.area,df.price,color = "r",marker = "*")
plt.title("home prices",fontsize = 15)
plt.xlabel("area",fontsize = 12)
plt.ylabel("price",fontsize = 12)
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[4000]])
```

```python
data = pd.read_csv("areas.csv")
data
```

```python
reg.predict(data)
```

```python
pre = reg.predict(data)
data["price"] = pre
data.to_csv("prediction.csv")
```

```python
%matplotlib inline
plt.scatter(df.area,df.price,color = "r",marker = "*")
plt.title("predicted home price",fontsize = 15)
plt.xlabel("area",fontsize = 12)
plt.ylabel("price",fontsize = 12)
plt.plot(df.area,reg.predict(df[['area']]),color = "g")
plt.show()
```
